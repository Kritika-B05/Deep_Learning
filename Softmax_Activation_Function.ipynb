```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(z):
  exp_z = np.exp(z -  np.max(z)) # Numerical stability
  return exp_z / np.sum(exp_z)

# Example input
z= np.array([2, 1, 0])
softmax_output = softmax(z)

#Plot the Softmax output
plt.bar(["Class 1", "Class 2", "Class 3"], softmax_output, color=['blue', 'green', 'red'])

plt.xlabel("Classes")
plt.ylabel("Softmax Probability")
plt.title("Softmax Function Output")
plt.ylim(0, 1) # Set y-axis limit to [0,1]
plt.show()
```
